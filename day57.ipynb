```python
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('/content/sample_data/cars.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type'] = df['Fuel_Type'].map({'Petrol':0, 'Diesel':1})
```

```python
df['Transmission'].unique()
```

```python
df['Transmission'] = df['Transmission'].map({'Manual':0, 'Automatic':1})
```

```python
df['Owner_Type'].unique()
```

```python
df['Owner_Type'] = df['Owner_Type'].map({'First':0, 'Second':1, 'Third':2})
```

```python
df = df.drop(columns=['Car_ID', 'Brand', 'Model'])
```

```python
df.head(3)
```

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, random_state=92)
```

```python
tf.random.set_seed(48)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation= 'relu'),
    tf.keras.layers.Dense(1, activation ='relu')
])

model.compile( loss = tf.keras.losses.MeanSquaredLogarithmicError,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.1))

model.fit(X_train, y_train, epochs=100,verbose = 1)
```

```python
model.evaluate(X_test, y_test)
```
