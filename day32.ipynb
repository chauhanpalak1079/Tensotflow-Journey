```python
import tensorflow as tf
import pandas as pd
```

```python
from sklearn.datasets import make_circles
c = 3000
A, b = make_circles(c, noise=0.03, random_state=9)
```

```python
import sklearn
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(A,b,test_size=0.3,random_state=42)
```

```python
tf.random.set_seed(42)

model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation = 'relu'),
    tf.keras.layers.Dense(66, activation = 'relu'),
    tf.keras.layers.Dense(33, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model1.compile( loss = tf.keras.losses.BinaryCrossentropy, optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), metrics = ['accuracy'] )

model1.fit(x_train, y_train, epochs = 100, verbose=0, validation_data=(x_test, y_test))

model1_evaluation = model1.evaluate(x_test, y_test)
print(f"model 1 evaluation: {model1_evaluation}")
```

```python
tf.random.set_seed(42)

model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(55, activation = 'sigmoid'),
    tf.keras.layers.Dense(44, activation = 'sigmoid'),
    tf.keras.layers.Dense(33, activation = 'sigmoid'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model2.compile( loss = tf.keras.losses.BinaryCrossentropy, optimizer = tf.keras.optimizers.RMSprop(learning_rate= 0.01), metrics = ['accuracy'] )

model2.fit(x_train, y_train, epochs = 100, verbose=0)

model2_evaluation = model2.evaluate(x_test, y_test)
print(f"model 2 evaluation: {model2_evaluation}")
```

```python
tf.random.set_seed(42)

model3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation = 'leaky_relu'),
    tf.keras.layers.Dense(50, activation = 'leaky_relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model3.compile( loss = tf.keras.losses.BinaryCrossentropy, optimizer = tf.keras.optimizers.Adagrad(learning_rate= 0.01), metrics = ['accuracy'] )

model3.fit(x_train, y_train, epochs = 100, verbose=0, validation_data=(x_test, y_test))

model3_evaluation = model3.evaluate(x_test, y_test)
print(f"model 3 evaluation: {model3_evaluation}")
```

```python
tf.random.set_seed(42)

model4 = tf.keras.Sequential([
    tf.keras.layers.Dense(200, activation = 'relu'),
    tf.keras.layers.Dense(100, activation = 'relu'),
    tf.keras.layers.Dense(50, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model4.compile( loss = tf.keras.losses.BinaryCrossentropy, optimizer = tf.keras.optimizers.SGD(learning_rate= 0.01), metrics = ['accuracy'] )

model4.fit(x_train, y_train, epochs = 100, verbose=0, validation_data=(x_test, y_test))

model4_evaluation = model4.evaluate(x_test, y_test)
print(f"model 4 evaluation: {model4_evaluation}")
```

```python
tf.random.set_seed(42)

model5 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation = 'softmax'),
    tf.keras.layers.Dense(70, activation = 'softmax'),
    tf.keras.layers.Dense(30, activation = 'softmax'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model5.compile( loss = tf.keras.losses.BinaryCrossentropy, optimizer = tf.keras.optimizers.Adam(learning_rate= 0.01), metrics = ['accuracy'] )

model5.fit(x_train, y_train, epochs = 100, verbose=0, validation_data=(x_test, y_test))

model5_evaluation = model5.evaluate(x_test, y_test)
print(f"model 5 evaluation: {model5_evaluation}")
```

```python
print("Summary of models: ")
print(f"Model 1: {model1_evaluation}")
print(f"Model 2: {model2_evaluation}")
print(f"Model 3: {model3_evaluation}")
print(f"Model 4: {model4_evaluation}")
print(f"Model 5: {model5_evaluation}")
```
