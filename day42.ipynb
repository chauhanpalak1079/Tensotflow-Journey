```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/kc_house_data.csv')
```

```python
df.head(3)
```

```python
df = df.drop(columns = ["id","date","zipcode","lat","long"])
```

```python
df.info()
```

```python
df.head(3)
```

```python
y = df['price']
x = df.drop('price',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5,random_state=55)
```

```python
tf.random.set_seed(100)

model=tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation = 'leaky_relu'),
    tf.keras.layers.Dense(70, activation = 'leaky_relu'),
    tf.keras.layers.Dense(25, activation = 'leaky_relu'),
    tf.keras.layers.Dense(25, activation = 'leaky_relu'),
    tf.keras.layers.Dense(5, activation= 'leaky_relu')])

model.compile(loss = tf.keras.losses.MeanAbsolutePercentageError(),
              optimizer = tf.keras.optimizers.Adamax(learning_rate = 0.01),
              metrics =['mae'])
model.fit(x_train, y_train, epochs =100, verbose =1 , validation_data = (x_test, y_test))
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```
