```python
import tensorflow as tf
```

```python
data = tf.cast(tf.range(-30,30), tf.float32)
```

```python
data
```

```python
def sigmoid(x):
  return 1/(1+tf.exp(-x))
```

```python
def relu(x):
  return tf.maximum(0,x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(relu(data))
```

```python
plt.plot(sigmoid(data))
```

```python
def tanh(x):
  return (tf.exp(x)-tf.exp(-x))/(tf.exp(x)+tf.exp(-x))
```

```python
tanh(data)
```

```python
plt.plot(tanh(data))
```
