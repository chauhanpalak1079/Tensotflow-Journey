```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_circles
c = 1000
A, b = make_circles(c, noise=0.02, random_state=7)
```

```python
circles = pd.DataFrame({'X0': A[:,0],
                        'X1': A[:,1],
                        'label': b})
circles
```

```python
plt.scatter(circles['X0'][circles['label'] == 0], circles['X1'][circles['label'] == 0], color='green')
plt.scatter(circles['X0'][circles['label'] == 1], circles['X1'][circles['label'] == 1], color='red')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()
```
